```python
t_AMPA = 2
t_NMDA  = 50
t_GABA = 8

alpha = 1
def numerical_integrations(d_t, s0=1,alpha=1,r=10,T=10,tau=t_AMPA):
    list_s = []
    ''' numerical integration with different dt values'''
    s = s0
    list_s = [s0]
    time = np.arange(0,T, d_t)
    i=1
    for t in time[1:len(time)]:
        if ( t%(1000/r)==0):
            s = 1 + d_t*(alpha - s/tau)
        else :
            s= s + d_t*(- s/tau)
        i=i+1
        list_s += [s]
    return (list_s)
```

```python
plt.plot(np.arange(0,1000, 0.01),numerical_integrations(d_t=0.01, s0=1,alpha=1,r=10,T=1000,tau=t_AMPA))

plt.title(" periodic train at frequency r = 10",fontsize=20)
plt.xlabel('Time',fontsize=20)
plt.ylabel('s',fontsize=20)
plt.show()

print ('The average of s for r = 10 is : {0:.4f} '.format(np.sum(numerical_integrations(d_t=0.01, s0=1,alpha=1,r=10,T=10000,tau=t_AMPA))/len(np.arange(0,10000, 0.01))))
plt.plot(np.arange(0,1000, 0.01),numerical_integrations(d_t=0.01, s0=1,alpha=1,r=50,T=1000,tau=t_AMPA))


plt.title(" periodic train at frequency r = 50",fontsize=20)
plt.xlabel('Time',fontsize=20)
plt.ylabel('s',fontsize=20)
plt.show()
print ('The average of s for r = 50 is : {0:.4f} '.format(np.sum(numerical_integrations(d_t=0.01, s0=1,alpha=1,r=50,T=10000,tau=t_AMPA))/len(np.arange(0,10000, 0.01))))
```

# Question 2:
# a) Time constant for a given firing rate r
#### $\frac{ds}{dt} =\alpha_s r (1-s) -\frac{s}{\tau_s}$
#### Let's find the time constant :
#### $\frac{ds}{dt} =\alpha_s r - \alpha_s r s - s \frac{1}{\tau_s}= \alpha_s r - s ( \alpha_s r + \frac{1}{\tau_s} )= \alpha_s r - s ( \frac{\alpha_s r \tau_s + 1} {\tau_s} )$.

#### Therefore, the time constant is : $ (\frac{\alpha_s r \tau_s + 1} {\tau_s})^{-1} $, i.e $\boxed{\frac {\tau_s}{\alpha_s r \tau_s + 1} }$

```python
t_AMPA = 2e-3
t_NMDA  = 50e-3
t_GABA = 8e-3
r = 20
alpha = 1

def time_constant_value(t,r=20,alpha=1):
    return (t/(alpha*r*t + 1))

print ('Time constant for AMPA = {0:.4f} '.format(1000*time_constant_value(t_AMPA,r=r,alpha=1)))
print ('Time constant for NMDA = {0:.4f} '.format(1000*time_constant_value(t_NMDA,r=r,alpha=1)))
print ('Time constant for GABAa = {0:.4f} '.format(1000*time_constant_value(t_GABA,r=r,alpha=1)))
```

# b) Steady state

#### $\frac{ds}{dt} =\alpha_s r (1-s) -\frac{s}{\tau_s}=0$
#### $\Rightarrow \alpha_s r - \alpha_s r s - s \frac{1}{\tau_s}=0$
#### $\Rightarrow \alpha_s r = s( \alpha_s r +  \frac{1}{\tau_s})$
#### $\Rightarrow \tau_s \alpha_s r = s( \tau_s \alpha_s r +  1)$
#### $\Rightarrow \boxed{s_{ss} = \frac{\tau_s \alpha_s r}{\tau_s \alpha_s r +  1}}$

```python
t_AMPA = 2e-3
t_NMDA  = 50e-3
t_GABA = 8e-3
r = 20
alpha = 1

def steady_state(t,r,alpha=1):
    return((t*alpha*r)/(1+t*alpha*r))
```

```python
ss_AMPA =[steady_state(t=t_AMPA,r=r,alpha=1) for r in np.linspace(0,100,10000)]
ss_NMDA =[steady_state(t=t_NMDA,r=r,alpha=1) for r in np.linspace(0,100,10000)]
ss_GABA =[steady_state(t=t_GABA,r=r,alpha=1) for r in np.linspace(0,100,10000)]
plt.plot(np.linspace(0,100,10000),ss_AMPA,label="AMPA")
plt.plot(np.linspace(0,100,10000),ss_NMDA,label="NMDA")
plt.plot(np.linspace(0,100,10000),ss_GABA,label="GABA")
plt.legend(fontsize=20)

plt.title("Steady state as function of r between 0 and 100 Hz, for the three different types of synaptic current",fontsize=20)
plt.xlabel('Firing rate r',fontsize=20)
plt.ylabel('Steady state',fontsize=20)
plt.show()
```

### * General Behavior for any $\tau$:

#### For the three curves, we start with a linear behavior of rate $\alpha \tau$, where $\tau$ is specific to the synaptic current $\tau_{AMPA}, \tau_{NMDA}, \tau_{GABA}$ because at the beginning the firing rate is very small
#### and therefore $\alpha \tau r << 1$ . Hence, $s_{ss}(r) \sim \alpha \tau r$. The synaptic response will be linear to the firing rate for very small values of r.
#### On the other hand, when $\alpha \tau r$ is very large (much bigger than 1), we can omit  the term (+1) in the denominator and therefore s saturates to the value 1.
#### Basically, the expected behavior overtime is to start with a linear curve of rate $\alpha \tau$ that will start to become more and more flat saturating to 1. 




### * How the $\tau$ values affect the behavior of the curves :

#### When r starts increasing. , we have to consider the different values of $\tau$:

#### 1- For AMPA , we have : $\tau_{AMPA}=$2e-3 . Therefore, $\frac{1}{\alpha \tau_{AMPA}}=500  Hz$ .Since r is plotted between 0 and 100 , we have : $r << \frac{1}{\alpha \tau_{AMPA}}$ and therefore $s_{ss} \sim \alpha \tau_{AMPA} r$ .
#### Therefore, the steady state for AMPA is almost linear in r with a rate close to $\alpha \tau_{AMPA}=$2e-3

#### 2- For NMDA , we have : $\tau_{NMDA}=$50e-3 . Therefore, $\frac{1}{\alpha \tau_{NMDA}}=20 Hz$ .  
#### Since r is plotted between 0 and 100 , we can not omit the (+1) term and the linear behavior is not true . The curve will start flattening more and more and saturate to 1 .

#### 3- For GABA , we have : $\tau_{GABA}=$8e-3 . Therefore, $\frac{1}{\alpha \tau_{NMDA}}=125 Hz$ .  
#### Since r is plotted between 0 and 100 , we have 2 behaviors: for small values of r , we can still consider that the steady state is linear in r but as r increase , it does not hold anymore and the curve will start flattening more and more.

#### This has physiological implications because it shows that there is only a finite amount of receptor-channel complex that can be activated at a specific synapse.
#### For low firing rates and for AMPA , the synaptic response is linear with the firing rate. For large values of firing rate, the synaptic response will be flattening and then plateau to 1 (we can make analogies with the chemistry dosage curves).

#### This is also fundamental when we think about the fact that synaptic transmission is probabilistic and $\alpha$ is a dynamical variable and therefore the linear rate changes with respect to the  probability and the number of transmitors.
