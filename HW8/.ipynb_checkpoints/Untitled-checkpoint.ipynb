# Let's prepare the input vectors and output vectors with N=100 and p=10:

```python
N = 100 # i.e input vector size
p = 10 # i.e number of data

# Data : Input-Output generation :
random.seed(17)
X = np.array([[random.choice([-1, 1]) for i in range(N)] for j in range(p)])
Y = np.array([random.choice([-1, 1]) for j in range(p)])

# initialise weights and bias :
np.random.seed(17)
W_j = [np.random.normal(0,1) for i in range(N)]
theta = np.random.normal(0,1)

prop_errors = []

epoch = 0
errors_to_add = 0
for epochs in range(100) :
    errors_to_add = 0
    for t in range(p) :
        y_hat = np.dot(W_j,X[t]) - theta
        if (np.sign(y_hat)!=Y[t]):
            W_j = W_j + Y[t] * X[t]
            theta = theta - Y[t]
            errors_to_add+=1
    prop_errors.append(100*errors_to_add/p)
```

```python
N = 100 # i.e input vector size
p = 10 # i.e number of data

# Data : Input-Output generation :
random.seed(17)
X = np.array([[random.choice([-1, 1]) for i in range(N)] for j in range(p)])
Y = np.array([random.choice([-1, 1]) for j in range(p)])

# initialise weights and bias :
np.random.seed(17)
W_j = [np.random.normal(0,1) for i in range(N)]
theta = np.random.normal(0,1)

prop_errors = []

epoch = 0
errors_to_add = 0
#for epochs in range(100) :
errors_to_add = 0
for i in range(p*100) : # 100 times number of patterns
    random.seed(i)
    chosen_pattern = random.randint(0,p-1) # chose randomly one of the 10 patterns
    y_hat = np.dot(W_j,X[chosen_pattern]) - theta
    if (np.sign(y_hat)!=Y[chosen_pattern]):
        W_j = W_j + Y[chosen_pattern] * X[chosen_pattern]
        theta = theta - Y[chosen_pattern]
        errors_to_add+=1
    prop_errors.append(100*errors_to_add/p)
```

## $\frac{ \partial l} {\partial w_2} = 0 $
